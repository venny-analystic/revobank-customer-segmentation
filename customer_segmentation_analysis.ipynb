**Data Dictionary**

***Transaction Data :***
* **id** = Unique identifier for each transaction record.
* **date**	= The date when the transaction happened.
* **clien_id**	= ID of the user who made the transaction (linked to USER DATA).
* **card_id**	= ID of the card used in the transaction (linked to CARD DATA).
* **amount**	= The amount of money spent in the transaction.
* **user_id**	= Describes how the transaction was made:
  - Chip Transaction: The card's chip was used (more secure).
  - Swipe Transaction: The magnetic stripe was swiped (less secure).
  - Online Transaction: Made over the internet (card not physically used).
* **merchant_id**	= Unique identifier of the merchant (store / vendor) where the purchase happened.
* **merchant_city** =	City where the merchant is located.
* **merchant_state**	= State / region code of the merchant.
* **postcode**	= Postal code (ZIP code) of the merchant's location.
* **mcc** =	Merchant Category Code - a number that shows what kind of product/service was purchased.
* **errors**	= "Lists the issues that occurred during a transaction (if any). Can be one or multiple errors separated by commas:
  - Insufficient Balance: Not enough money or credit limit in the account.
  - Bad Expiration: The card was expired.
  - Bad CVV: The security code (CVV) was incorrect.
  - Bad PIN: The PIN was incorrect.
  - Bad Zipcode: The entered ZIP code didn't match the billing info.
  - Bad Card Number: The card number was invalid.
  - Technical Glitch: A system or network error occurred.
Can be empty (NaN) if no error occurred."

***MCC Code Data :***
* **mcc** = Code that represents the type of business or product (used in mcc field in TRANSACTION DATA).
* **mcc_description**	= A text label explaining what the merchant sells (e.g., "Restaurant", "Electronics")

***Train Fraud Label Data :***
* **transaction_id** = The transaction identifier (linked to id in TRANSACTION DATA).
* **target**	= Whether the transaction was fraud (True = fraud, False = Not fraud).

##a) Calculate Net Profit & MDR

```python
df_merged.info()
df_merged.describe()
df_merged.isnull().sum()
```

```python
#menghitung Net Profit (MDR)
money_cols = ['amt_nonfraud_trx_L6M', 'amt_fraud_trx_L6M']

for col in money_cols:
    df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce')

total_nonfraud_amt = df_merged['amt_nonfraud_trx_L6M'].sum()
total_fraud_amt = df_merged['amt_fraud_trx_L6M'].sum()

MDR_RATE = 0.015
net_profit = (total_nonfraud_amt * MDR_RATE) - total_fraud_amt

net_profit
```

```python
print("Net Profit (last 6 months):", net_profit)
```

Insight : The calculated net profit indicates that the banks revenue from transaction fees is sufficient to absorb fraud-related losses. This suggests that fraud risk, while present, does not materially threaten overall transaction profitability.

##b) Fraud Rate Pie Chart

```python
#Menghitung Fraud rate
total_transaction_amt = total_nonfraud_amt + total_fraud_amt
fraud_rate = total_fraud_amt / total_transaction_amt

fraud_rate
```

```python
#data Visual pie chart
import matplotlib.pyplot as plt

values = [total_nonfraud_amt, total_fraud_amt]
labels = ['Non-Fraud Transactions', 'Fraud Transactions']

plt.figure()
plt.pie(values, labels=labels, autopct='%1.2f%%', startangle=90)
plt.title('Fraud vs Non-Fraud Transactions (Based on Amount)')
plt.show()
```

Insight : The proportion of fraud transaction value is very small compared to total transaction value, suggesting that fraud exposure remains limited and does not significantly impact overall transaction performance.

##c) Different in transaction behavior percard brand use bar chart

```python
#analisis percard_brand total transaksi (count), total fraud transaksi (count), fraud rate, total nominal transaksi non fraud
df_merged['card_brand'] = df_merged['card_brand'].str.upper()

brand_behavior = df_merged.groupby('card_brand').agg(
    total_nonfraud_amt=('amt_nonfraud_trx_L6M','sum'),
    total_fraud_amt=('amt_fraud_trx_L6M','sum')
).reset_index()

brand_behavior
```

insight : The distribution of transaction amounts indicates that Mastercard and Visa dominate transaction activity, suggesting higher customer adoption or usage frequency. In contrast, Amex and JCB represent a smaller share of total transaction value, indicating lower transaction penetration in the observed period. Mastercard and Visa account for the majority of transaction volume, while Amex and JCB contribute relatively smaller portions.
